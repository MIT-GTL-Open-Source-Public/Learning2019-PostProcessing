# Plotting teams exploration of tradespace over the theoretical traedspace

First making sure that I can read the tradespace enumeration files generated from MATLAB.


```python
import os.path
from scipy.io import loadmat

ANALYTICAL_TRADESPAVCE_FILE = os.path.expanduser("~/Learning2019Data/AnalyticalTradespace/tradespaceDataVersion2.mat")
x = loadmat(ANALYTICAL_TRADESPAVCE_FILE)

timeCampus       = list()
totalCost        = list()
interactionScore = list()
diversityScore   = list()

for data_row in x['MasieResults']:
    for i in range(len(data_row)):
        timeCampus.append       (data_row[i][0][0][ 9][0][0])
        totalCost.append        (data_row[i][0][0][10][0][0])
        interactionScore.append (data_row[i][0][0][11][0][0])
        diversityScore.append   (data_row[i][0][0][12][0][0])
```

Now plotting the theoretical tradespace data only to check for consistency with MATLAB plots

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 22}

matplotlib.rc('font', **font)

GOLDEN_RATIO = np.array([1.61803398875, 1])
plt.figure(figsize=GOLDEN_RATIO*8)

plt.plot(totalCost, interactionScore, marker="o", linewidth=0, markersize=15, fillstyle='none', alpha=0.15)
plt.xlabel ('Total Cost (millions of USD)')
plt.ylabel ('Interaction Score (relative)')
plt.show()
```

```python
import sys
sys.path.append("../CommonModules") # go to parent dir/CommonModules

import Learning2019GTL.Globals as Globals
import Learning2019GTL.DataConnector as DataConnector

teams_to_plot = [33, 15, 21]

plt.figure(figsize=GOLDEN_RATIO*5)
plt.plot(totalCost, np.array(interactionScore)*100, fillstyle='none', 
         marker="o", linewidth=0, markersize=5, alpha=0.5, color='#7f7f7f',
        label='computational')

for t_plot in teams_to_plot:
    team = DataConnector.Team()
    ui_events = team.getSimulationUIEventsByTeamID(t_plot)
    ui_events_valid = team.getSimulationUIEventsByTeamIDBetweenDensities(t_plot, [100, 500])

    team_cost        = list()
    team_interaction = list()
    team_cost_valid        = list()
    team_interaction_valid = list()
    team_diversity_valid   = list()
    team_walkingtime_valid = list()
    for event in ui_events:
        team_cost.append(event.total_cost)
        team_interaction.append(event.interaction_score)
    
    for event in ui_events_valid:
        team_cost_valid.append(event.total_cost)
        team_interaction_valid.append(event.interaction_score)
        team_diversity_valid.append(event.diversity_score)
        team_walkingtime_valid.append(event.walking_time)
    
    team_cost        = np.array(team_cost)
    team_interaction = np.array(team_interaction)
    

#     plt.plot(team_cost, team_interaction, marker="o", linewidth=1, 
#               markersize=10, alpha=0.5, label=f'Team {t_plot}')
    plt.plot(team_cost_valid, team_interaction_valid, marker="o", linewidth=1, 
             fillstyle='none', markersize=10, alpha=0.75, label=f'Team {t_plot} Valid')
    

plt.legend(fontsize=12, loc='upper right', bbox_to_anchor=(1,1))
plt.xlabel ('Total Cost (millions of USD)')
plt.ylabel ('Interaction Score (relative)')
plt.show()

plt.figure(figsize=GOLDEN_RATIO*4)
plt.plot(totalCost, diversityScore, marker="o", linewidth=0, markersize=5, alpha=0.1, color='#7f7f7f')

for t_plot in teams_to_plot:
    team = DataConnector.Team()
    ui_events = team.getSimulationUIEventsByTeamID(t_plot)

    team_cost      = list()
    team_diversity = list()
    for event in ui_events:
        team_cost.append(event.total_cost)
        team_diversity.append(event.diversity_score)

    team_cost        = np.array(team_cost)
    team_diversity   = np.array(team_diversity)

    plt.plot(team_cost, team_diversity/100, marker="o", linewidth=1, 
             markersize=8, alpha=0.5, label=f'Team {t_plot}')

plt.legend(fontsize=16, loc='upper right', bbox_to_anchor=(1,1))
plt.xlabel ('Total Cost (millions of USD)')
plt.ylabel ('Diversity Score (relative)')
plt.show()

plt.figure(figsize=GOLDEN_RATIO*4)
plt.plot(totalCost, timeCampus, marker="o", linewidth=0, markersize=10, alpha=0.15, color='#7f7f7f')

for t_plot in teams_to_plot:
    team = DataConnector.Team()
    ui_events = team.getSimulationUIEventsByTeamID(t_plot)

    team_cost        = list()
    team_walkingtime = list()
    for event in ui_events:
        team_cost.append(event.total_cost)
        team_walkingtime.append(event.walking_time)

    team_cost        = np.array(team_cost)
    team_walkingtime = np.array(team_walkingtime)

    plt.plot(team_cost, team_walkingtime, marker="o", linewidth=1, 
             markersize=15, alpha=0.5, label=f'Team {t_plot}')

plt.legend(fontsize=16, loc='lower right', bbox_to_anchor=(1,0))
plt.xlabel ('Total Cost (millions of USD)')
plt.ylabel ('Walking Time (minutes)')
plt.show()
```
