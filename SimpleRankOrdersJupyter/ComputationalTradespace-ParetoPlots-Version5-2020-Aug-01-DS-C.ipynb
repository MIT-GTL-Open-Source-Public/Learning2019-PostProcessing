# Plotting teams exploration of tradespace over the theoretical traedspace

```python
import os.path
from scipy.io import loadmat
import csv
import pandas

ANALYTICAL_TRADESPACE_FILE = os.path.expanduser("~/Learning2019Data/AnalyticalTradespace/data-2020-Jul-29_filtered.csv")
PARETO_RANK_1 = [15, 303541, 330786, 17199, 539103, 300703, 198554, 520370, 526191, 195651, 170788, 195647, 225896]

df = pandas.read_csv(ANALYTICAL_TRADESPACE_FILE)
df_p1 = df.loc[df['ID'].isin(PARETO_RANK_1)]
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 22}

matplotlib.rc('font', **font)

df_p1 = df_p1.sort_values('cost')
#df_p2 = df_p2.sort_values('cost')


GOLDEN_RATIO = np.array([1.61803398875, 1])
plt.figure(figsize=GOLDEN_RATIO*8)

plt.plot(df['cost']/1e6, df['interaction_score'], marker="o", linewidth=0, 
         markersize=5, fillstyle='full', alpha=0.5)
plt.plot(df_p1['cost']/1e6, df_p1['interaction_score'], marker="o", linewidth=3, 
         markersize=5, fillstyle='full', alpha=1.0)
# plt.plot(df_p2['cost']/1e6, df_p2['interaction_score'], marker="o", linewidth=3, 
#          markersize=5, fillstyle='full', alpha=1.0)

plt.xlabel ('Total Cost (millions of USD)')
plt.ylabel ('Interaction Score (relative)')
plt.show()
```

```python
plt.figure(figsize=GOLDEN_RATIO*8)

plt.plot(df['cost']/1e6, df['team_mix'], marker="o", linewidth=0, 
         markersize=5, fillstyle='full', alpha=0.5)
plt.plot(df_p1['cost']/1e6, df_p1['team_mix'], marker="o", linewidth=3, 
         markersize=5, fillstyle='full', alpha=1.0)
# plt.plot(df_p2['cost']/1e6, df_p2['team_mix'], marker="o", linewidth=3, 
#          markersize=5, fillstyle='full', alpha=1.0)
plt.xlabel ('Total Cost (millions of USD)')
plt.ylabel ('Diversity Score (relative)')
plt.show()
```

```python
plt.figure(figsize=GOLDEN_RATIO*8)

plt.plot(df['cost']/1e6, df['walking_time'], marker="o", linewidth=0, 
         markersize=5, fillstyle='full', alpha=0.5)
plt.plot(df_p1['cost']/1e6, df_p1['walking_time'], marker="o", linewidth=3, 
         markersize=5, fillstyle='full', alpha=1.0)
# plt.plot(df_p2['cost']/1e6, df_p2['walking_time'], marker="o", linewidth=3, 
#          markersize=5, fillstyle='full', alpha=1.0)
plt.xlabel ('Total Cost (millions of USD)')
plt.ylabel ('Walking Time (minutes)')
plt.show()
```
